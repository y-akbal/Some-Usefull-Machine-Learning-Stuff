## Some gradient descent stuff ---

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def f(x):
    return x*x*x-5*x+10  
#### this guy has two local minimas, 
#### the gradient descent algorithm does not tell you which one
### it converges to !!!!!!!
```

```python
plt.plot(np.arange(-8, 8, 0.1), f(np.arange(-8, 8, 0.1)))
```

```python
def Df(x):
    return 3*x**2-5
```

```python
x_0 = np.random.randn()
```

```python
x_0
```

```python
L = []
lr = 0.001 
for i in range(1000):
    x_0 -= lr*Df(x_0)+##moment  ### x_0 = x_0 - alpha*Df(x_0)
    L.append(x_0)
```

```python
### A quick takeaway, you should choose a proper learning rate (lr) not too small (you learn pretty slow) 
### not too large (you lear pretty fast so that things diverge!)
```

```python
plt.plot(L)
```
